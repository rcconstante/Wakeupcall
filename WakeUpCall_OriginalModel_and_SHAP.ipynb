# Phase 1.5: Feature Engineering

```python
# 1. Create Neck_Above_40
df['Neck_Above_40'] = (df['Neck_Circumference'] > 40).astype(int)

# 2. Anthropometric ratios
df['Neck_Height_Ratio'] = df['Neck_Circumference'] / df['Height']
df['Weight_Height_Ratio'] = df['Weight'] / df['Height']

# 3. ESS Category (string)
def ess_category(score):
    if score < 10:
        return "Normal"
    elif 10 <= score <= 12:
        return "Mild Sleepiness"
    elif 13 <= score <= 15:
        return "Moderate Sleepiness"
    else:
        return "Severe Sleepiness"

df['ESS_Category'] = df['Epworth_Score'].apply(ess_category)

# 4. STOP Count
df['STOP_Count'] = (
    df['STOP_Snore'] +
    df['STOP_Tired'] +
    df['STOP_ObsApnea'] +
    df['STOP_Pressure']
)

# 5. BANG Total
df['BANG_Total'] = (
    df['BANG_BMI'] +
    df['BANG_Age'] +
    df['BANG_Neck'] +
    df['BANG_Gender']
)

# 6. Airway Composite
df['Airway_Composite'] = (
    0.3 * df['BMI'] +
    0.3 * df['Neck_Circumference'] +
    0.2 * df['Alcohol'] +
    0.2 * df['Snoring']
)

# 7. BMI Category
df['BMI_Category'] = pd.cut(
    df['BMI'],
    bins=[0, 25, 30, 35, 100],
    labels=[0, 1, 2, 3]
).astype(int)
```

# Phase 2: Data Preprocessing

```python
final_features = [

    # Demographics
    'Age', 'Sex', 'Age_Group',

    # Anthropometrics
    'Height', 'Weight', 'BMI', 'BMI_Category',
    'Neck_Circumference', 'Neck_Above_40',
    'Neck_Height_Ratio', 'Weight_Height_Ratio',

    # Lifestyle
    'Smokes', 'Alcohol',

    # Sleep symptoms
    'Snoring', 'Sleepiness', 'Epworth_Score', 'ESS_Category',

    # Medical history
    'Hypertension', 'Diabetes', 'Depression',

    # STOP components
    'STOP_Snore', 'STOP_Tired', 'STOP_ObsApnea', 'STOP_Pressure', 'STOP_Count',

    # BANG components
    'BANG_Age', 'BANG_BMI', 'BANG_Neck', 'BANG_Gender', 'BANG_Total',

    # Composite scores
    'STOPBANG', 'Berlin_Score',

    # Composite feature
    'Airway_Composite'
]

print("Number of final selected features:", len(final_features))
print("Final Features:\n", final_features)
```

```python
from sklearn.preprocessing import LabelEncoder

label_cols = ['ESS_Category']
le_dict = {}

for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    le_dict[col] = le

print("\nEncoded categorical columns:", label_cols)
```

```python
risk_map = {'High': 0, 'Intermediate': 1, 'Low': 2}
df['OSA_Risk'] = df['OSA_Risk'].map(risk_map)

print("\nMapped Target Distribution:")
print(df['OSA_Risk'].value_counts())
```

```python
X = df[final_features].copy()
y = df['OSA_Risk']

print("\nX Shape:", X.shape)
print("y Shape:", y.shape)
```

```python
print("\nMissing values in X:")
print(X.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

print("\nTraining set:", X_train.shape)
print("Testing set:", X_test.shape)
```

```python
numeric_cols = [
    'Age', 'Height', 'Weight', 'BMI', 'BMI_Category',
    'Neck_Circumference', 'Neck_Height_Ratio', 'Weight_Height_Ratio',
    'Epworth_Score', 'STOP_Count', 'BANG_Total', 'STOPBANG',
    'Berlin_Score', 'Airway_Composite'
]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])

print("\nScaling complete.")
```

# Phase 3: Train the Final LightGBM Model

```python
import lightgbm as lgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
final_lgb = lgb.LGBMClassifier(
    objective='multiclass',
    num_class=3,
    boosting_type='gbdt',

    num_leaves=15,
    max_depth=4,
    min_data_in_leaf=3,
    learning_rate=0.05,
    n_estimators=300,
    feature_fraction=0.9,
    bagging_fraction=0.8,
    bagging_freq=5,

    random_state=42
)

# Train on engineered dataset
final_lgb.fit(X_train, y_train)

print("Final LightGBM model trained!")
```

```python
# Predictions on test split
lgb_test_preds = final_lgb.predict(X_test)
```

```python
# Predictions on test split
lgb_test_preds = final_lgb.predict(X_test)

# Evaluation
print("\nðŸ”µ FINAL LIGHTGBM PERFORMANCE (Engineered Dataset)")
print("Accuracy:", accuracy_score(y_test, lgb_test_preds))

print("\nClassification Report:\n",
      classification_report(y_test, lgb_test_preds))

print("Confusion Matrix:\n",
      confusion_matrix(y_test, lgb_test_preds))
```

# Phase 4: SHAP Explainability

```python
import shap
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Restore raw values for dependence plots
X_train_raw = X_train.copy()
X_train_raw[numeric_cols] = scaler.inverse_transform(X_train[numeric_cols])

print("Raw data restored:", X_train_raw.shape)
```

```python
# Create explainer (multiclass LightGBM)
explainer = shap.TreeExplainer(final_lgb, model_output="raw")

# Compute SHAP values on SCALED features
shap_values_raw = explainer.shap_values(X_train)

print("SHAP values type:", type(shap_values_raw))
```

```python
# Convert SHAP format to: list of 3 arrays (one per class)
if isinstance(shap_values_raw, list):
    # Already a list of [class0, class1, class2]
    shap_values_list = shap_values_raw
else:
    # Convert (n_samples, n_features, n_classes) â†’ list
    shap_values_arr = np.array(shap_values_raw)
    shap_values_list = [
        shap_values_arr[:, :, i] for i in range(shap_values_arr.shape[2])
    ]

print("\nFinal SHAP list lengths:", len(shap_values_list))
for i, arr in enumerate(shap_values_list):
    print(f"Class {i} SHAP shape:", arr.shape)
```

```python
# Class names in correct order
class_names = ["High", "Intermediate", "Low"]

# Extract SHAP matrices per class
shap_high         = shap_values_list[0]   # Class 0
shap_intermediate = shap_values_list[1]   # Class 1
shap_low          = shap_values_list[2]   # Class 2

print("shap_high:", shap_high.shape)
print("shap_intermediate:", shap_intermediate.shape)
print("shap_low:", shap_low.shape)
```

**Global SHAP Summary (Beeswarm)**

```python
print("\nðŸ”µ Generating per-class SHAP beeswarm plots...")

# High-risk
print("\n[Class: High Risk]")
shap.summary_plot(
    shap_high,          # (n_samples, n_features)
    X_train,            # same shape (scaled features)
    feature_names=X_train.columns,
    plot_type='dot'
)

# Intermediate-risk
print("\n[Class: Intermediate Risk]")
shap.summary_plot(
    shap_intermediate,
    X_train,
    feature_names=X_train.columns,
    plot_type='dot'
)

# Low-risk
print("\n[Class: Low Risk]")
shap.summary_plot(
    shap_low,
    X_train,
    feature_names=X_train.columns,
    plot_type='dot'
)
```

**SHAP Summary Bar Plot (Mean Absolute)**

```python
print("\nðŸ”µ Generating SHAP Feature Importance (Bar Plot)...")

shap.summary_plot(
    shap_values_list,       # list of arrays
    X_train,
    feature_names=X_train.columns,
    class_names=class_names,
    plot_type='bar',
    max_display=20
)
```

**SHAP Dependence Plots**

```python
# Get global importance over all classes
shap_arr = np.array(shap_values_list)              # (3, n_samples, n_features)
mean_abs = np.mean(np.abs(shap_arr), axis=(0, 1))  # (n_features,)
sorted_idx = np.argsort(mean_abs)[::-1]
top_features = X_train.columns[sorted_idx][:5]

print("\nTop 5 SHAP Features:", list(top_features))
```

```python
for feature in top_features:
    print(f"\nðŸ”µ Low-Risk Dependence Plot â†’ {feature}")
    shap.dependence_plot(feature, shap_low, X_train, display_features=X_train_raw)

for feature in top_features:
    print(f"\nðŸŸ  Intermediate-Risk Dependence Plot â†’ {feature}")
    shap.dependence_plot(feature, shap_intermediate, X_train, display_features=X_train_raw)

for feature in top_features:
    print(f"\nðŸ”´ High-Risk Dependence Plot â†’ {feature}")
    shap.dependence_plot(feature, shap_high, X_train, display_features=X_train_raw)
```

# PHASE 6: Testing the Model with Sample Inputs

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder

le_risk = LabelEncoder()
le_risk.fit(y)   # y already contains 0,1,2

print("Risk classes learned:", le_risk.classes_)
```

```python
def prepare_user_input(raw_dict):

    df_input = pd.DataFrame([raw_dict]).copy()

    # ---- ENGINEERED FEATURES ----
    df_input["Neck_Above_40"] = (df_input["Neck_Circumference"] > 40).astype(int)
    df_input["Neck_Height_Ratio"] = df_input["Neck_Circumference"] / df_input["Height"]
    df_input["Weight_Height_Ratio"] = df_input["Weight"] / df_input["Height"]

    # ESS Category â†’ must be string before encoding
    def ess_category(score):
        if score < 10:
            return "Normal"
        elif score <= 12:
            return "Mild Sleepiness"
        elif score <= 15:
            return "Moderate Sleepiness"
        else:
            return "Severe Sleepiness"

    df_input["ESS_Category"] = df_input["Epworth_Score"].apply(ess_category)

    df_input["STOP_Count"] = (
        df_input["STOP_Snore"] +
        df_input["STOP_Tired"] +
        df_input["STOP_ObsApnea"] +
        df_input["STOP_Pressure"]
    )

    df_input["BANG_Total"] = (
        df_input["BANG_BMI"] +
        df_input["BANG_Age"] +
        df_input["BANG_Neck"] +
        df_input["BANG_Gender"]
    )

    df_input["Airway_Composite"] = (
        0.3 * df_input["BMI"] +
        0.3 * df_input["Neck_Circumference"] +
        0.2 * df_input["Alcohol"] +
        0.2 * df_input["Snoring"]
    )

    df_input["BMI_Category"] = pd.cut(
        df_input["BMI"],
        bins=[0, 25, 30, 35, 100],
        labels=[0, 1, 2, 3]
    ).astype(int)

    # ---- ENCODE ONLY ESS_CATEGORY ----
    for col, encoder in le_dict.items():
        df_input[col] = encoder.transform(df_input[col].astype(str))

    # ---- ENSURE COLUMN ORDER ----
    df_input = df_input[final_features]

    # ---- SCALE NUMERICAL FEATURES ----
    df_input[numeric_cols] = scaler.transform(df_input[numeric_cols])

    return df_input
```

```python
def predict_user(raw_dict):

    df_ready = prepare_user_input(raw_dict)

    # Predict encoded label (0, 1, 2)
    pred_encoded = final_lgb.predict(df_ready)[0]

    # Convert numeric label back to text class
    pred_label = le_risk.inverse_transform([pred_encoded])[0]

    # Probabilities of all 3 classes
    probs = final_lgb.predict_proba(df_ready)[0]

    # Certainty %
    certainty = float(probs[pred_encoded] * 100)

    return pred_label, probs, certainty
```

```python
def print_prediction(label_encoded, prob, certainty):
    risk_labels = {0: "High", 1: "Intermediate", 2: "Low"}
    readable = risk_labels[label_encoded]

    print(f"Prediction: {readable}")
    print(f"Probabilities: {prob}")
    print(f"Certainty: {certainty:.2f}%")
```
